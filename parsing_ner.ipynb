```python
import json
```

```python
data = json.load(open('./ner_inverted.json'))
```

```python
temp = {}
for word,doc in data.items():
    word = word.strip().lower()
    try:
        temp[word] = temp[word] or set(doc)
    except KeyError:
        temp[word] = set(doc)
```

```python
qq = temp.copy()
l = [',','.','u00a0','u2013','u2009','u2019','u201c','are ','and ','sunday,','monday,','tuesday,','wednesday,','thursday,','friday,','saturday,']
total = []
for word in qq:
    for check in l:
        if(check == word[:len(check)]):
            total.append(word)
            remove = word[len(check):].lower().strip()
            if(len(remove)):
                try:
                    temp[remove] = temp[word] or temp[remove]
                except:
                    temp[remove] = temp[word]
            del temp[word]
len(temp),total
```
